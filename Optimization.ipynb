```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.optimize as opt
```

OPTIMIZATION 

Is the branch of mathematics focused on finding extreme values (max or min) of functions

```python
def f(x):
  return x**4 - 3*4**2

def fp(x):
  return 4*x**3 - 6*x

# creating 100 evenly spaced points between -2 and 2
x = np.linspace(-2., 2., 100)

fx = f(x)
fpx = fp(x)

fig, ax = plt.subplots(1, 2)

ax[0].plot(x, fx)
ax[0].set_title("Function")

ax[1].plot(x, fpx)
ax[1].hlines(0.0, -2.5, 2.5, color="k", linestyle="--")
ax[1].set_title("Derivative")

for _ax in ax:
  _ax.spines["right"].set_visible(False)
  _ax.spines["top"].set_visible(False)
```
