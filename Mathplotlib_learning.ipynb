```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax =  plt.subplots()
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
ax.plot(x,y)
```

```python
fig, ax = plt.subplots()
fig.set_facecolor("red")
ax.set_facecolor("blue")
```

```python
### Multiple axis on a figure
fig, axes = plt.subplots(2, 3)
fig.set_facecolor("grey")
colors = ["#065535", "#89ecda", "#ffd1dc", "#ff0000","#9400d3", "#6897bb"]

for (ax, c) in zip(axes.flat, colors):
  ax.set_facecolor(c)

fig.tight_layout()
```

```python
countries = ["canada", "mexico", "america"]
populations = [36.7, 129.2, 325.700]
land_area = [3.850, 0.761, 3.790]

fig, ax = plt.subplots(2)

ax[0].bar(countries, populations, align= "center")
ax[0].set_title("Populations (in millions)")

ax[1].bar(countries, land_area, align= "center")
ax[1].set_title("Land area (in millions milies squared)")

fig.tight_layout()
```

```python

from matplotlib.patches import ArrowStyle
n = 50
np.random.seed(42)

x = np.random.rand(n)
y = np.random.rand(n)
colors = np.random.rand(n)
area = np.pi * (15 * np.random.rand(n))**2

fig, ax = plt.subplots()
ax.scatter(x, y, s=area, alpha=0.5)

ax.annotate(
    "first point", xy=(x[0], y[0]), xycoords="data",
    xytext=(25, -25), textcoords="offset points",
    arrowprops=dict(ArrowStyle="->", ConnectionStyle="arc3, rad=0")
    )
```

```python
x = np.linspace(0, 1, 500)
y = np.sin(4 * np.pi * x) * np.exp(-5 * x)

fig, ax = plt.subplots()
ax.grid(True)
ax.fill(x,y)
```
