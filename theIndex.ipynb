**SETTING THE INDEX**

```python
# CREATE the dataframe
df_year = df.set_index(["year"])
df_year.head()
```

```python
# Extract all data for a specific year...
df_year.loc[2015]
```

```python
# Example, if we wanted to find the difference in the average of all the variables from a year to another year.
df_year.loc[2010].mean(numeric_only=True) - df_year.loc[2009].mean(numeric_only=True)
```

```python
# Whats the Imports in Germany in 2012?
xyz = df_year.loc[df_year["country"] == "Germany", "Imports"].loc[2012]
print("Germany imports in 2012 is : ", xyz)
```

```python
# Whats the GDP in the USA and the UK in 2010?
abc = df_year.loc[df_year["country"].isin(["United States", "United Kingdom"]), "GDP"].loc[2010]
print("GDP in usa and uk in 2010 is : ")
abc
```

```python
df_year.loc[2010]
```

**SETTING A HIERARCHICAL INDEX**

Includes multiple columns in the index is advantageous in some situations...

Example, 
1. When we need more than one piece of information(column) to identify an abservation.
2. When we need data-alignment by more than one column.

To achieve multiple columns in the index, we pass a list of multiple column names to    `set_index`




```python
wdi = df.set_index(["country", "year"])
wdi.head(22)
```

# The row lables have 2 rows.
1. outer (left-most) row = country
2. inner (right-most) row = year 

* When a DataFrame's index has multiple levels, we refer to the dataFrame as having hierarchical index. 

**SLICING A HIERARCHICAL INDEX**

```python
# Answering the previous questions in different way

# Whats the Imports in Germany in 2012?
xyz = wdi.loc[("Germany", 2012), "Imports"]
#xyz = df_year.loc[df_year["country"] == "Germany", "Imports"].loc[2012]
print("Germany imports in 2012 is : ", xyz)

print("")

# Whats the GDP in the USA and the UK in 2010?
abc = wdi.loc[(["United States", "United Kingdnom"], 2010), "GDP"]
#abc = df_year.loc[df_year["country"].isin(["United States", "United Kingdom"]), "GDP"].loc[2010]
print("GDP in usa and uk in 2010 is : \n")
abc
```
