```python
url = "https://datascience.quantecon.org/assets/data/wdi_data.csv"
df = pd.read_csv(url)
df.info()
print("======= --- =======")
df
```

** Automatic Alignment. **

=> df_selected.loc[row, col] + imp_exp.loc[row, col]




**Handling Missing Data**

Example, for all elements in row 1 OR columns ["country", "year", "GovExpend", "Consumption", "GDP"], the value in selected_impExp is NaN

This is the way Pandas represents missing data. 

So, when pandas trying to look up the values in the table and could only find one vlaue in the dataFrame and other one was missing, 
then it says that the result is missing (NaN)

**SETTING THE INDEX**

```python
# CREATE the dataframe
df_year = df.set_index(["year"])
df_year.head()
```

```python
# Extract all data for a specific year...
df_year.loc[2015]
```

```python
# Example, if we wanted to find the difference in the average of all the variables from a year to another year.
df_year.loc[2010].mean(numeric_only=True) - df_year.loc[2009].mean(numeric_only=True)
```

```python
# Whats the Imports in Germany in 2012?
xyz = df_year.loc[df_year["country"] == "Germany", "Imports"].loc[2012]
print("Germany imports in 2012 is : ", xyz)
```

```python
# Whats the GDP in the USA and the UK in 2010?
abc = df_year.loc[df_year["country"].isin(["United States", "United Kingdom"]), "GDP"].loc[2010]
print("GDP in usa and uk in 2010 is : ")
abc
```
